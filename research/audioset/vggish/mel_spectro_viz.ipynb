## Import needed libraries

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
%matplotlib widget
```

## Loading file with librosa

```python
my_file = "Data/great_tit.wav"
Audio(my_file)
```

```python
# Load the audio file int librosa
scale, sr = librosa.load(my_file)
```

## Mel filter banks

```python
filter_banks = librosa.filters.mel(n_fft=2048, sr=sr, n_mels=10)
```

```python
filter_banks.shape
```

```python
plt.figure(figsize=(10,4))
librosa.display.specshow(filter_banks,
                         sr=sr,
                         x_axis='linear')
plt.colorbar(format="%+2.f")
```

## Extract Mel spectrogram

```python
mel_spectro = librosa.feature.melspectrogram(y=scale, sr=sr, n_fft=2048, hop_length=512, n_mels=90)
mel_spectro.shape
```

```python
log_mel_spectro = librosa.power_to_db(mel_spectro)
```

```python
plt.figure(figsize=(10,4))
librosa.display.specshow(log_mel_spectro,
                         x_axis="time",
                         y_axis="mel",
                         sr=sr)
plt.colorbar(format="%+2.f")
```
